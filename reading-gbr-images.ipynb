```python
#Reading the image array
import cv2
path ='/kaggle/input/tensorflow-great-barrier-reef/train_images/video_1/6234.jpg'
img = cv2.imread(path)
```

```python
#How do I convert a numpy array to (and display) an image:
from matplotlib import pyplot as plt
plt.imshow(img, interpolation='nearest')
plt.show()
```

```python
#Reading the train.csv file
import pandas as pd
train_df = pd.read_csv('/kaggle/input/tensorflow-great-barrier-reef/train.csv')
```

```python
# Quick view of my train_df dataframe
print(train_df.tail())
train_df.dtypes
```

```python
# Filtering the information of our desired image
train_df.loc[( train_df['video_frame'] == 6234) & (train_df['video_id'] == 1)]
```

```python
# Drawing the boundary box on the image as per the given annotations
import matplotlib.patches as patches

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(img)

# Create a Rectangle patch
rect = patches.Rectangle((566, 33), 57, 57, linewidth=1, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
